```python
import pandas as pd

from sf_permits.config import INTERIM_DATASET_PATH
```

* Units and stories with respect to location
* Repeated permit number and street names
* How many places have the same location
    * Consider not only coordinates
    * Street name and number

```python
df = pd.read_parquet(INTERIM_DATASET_PATH)
df.shape
```

```python
LOCATION_IDENITFIERS = [
    "Permit Number",
    "Street Name",
    "Street Number",
    "Street Number Suffix",
    "Unit",
    "Unit Suffix",
]
```

```python
location_count = df[LOCATION_IDENITFIERS].value_counts(dropna=False)
location_count
```

```python
location_count[location_count > 1]
```

```python
unique_df = df.groupby(LOCATION_IDENITFIERS, dropna=False).first().reset_index()
unique_df.shape
```

```python
unique_df.shape[0] / df.shape[0]
```
