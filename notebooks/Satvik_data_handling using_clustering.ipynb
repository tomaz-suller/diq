```python
import re

def clean_description(text):
    if pd.isnull(text):
        return ""
    text = text.lower()                     # Convert to lowercase
    text = re.sub(r'\d+', '', text)         # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)     # Remove punctuation
    text = re.sub(r'\s+', ' ', text)        # Replace multiple spaces with a single space
    return text.strip()                     # Strip leading/trailing spaces

permits["Cleaned_Description"] = permits["Description"].apply(clean_description)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

def remove_stopwords(text):
    words = text.split()
    return " ".join(word for word in words if word not in stop_words)

permits["Cleaned_Description"] = permits["Cleaned_Description"].apply(remove_stopwords)
```

### Transformed the cleaned descriptions into numerical representations using TF-IDF (Term Frequency-Inverse Document Frequency). This approach assigns higher importance to unique words in each document while reducing the weight of common words across all documents.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Vectorize text
vectorizer = TfidfVectorizer(max_features=1000)  # Limit features to reduce noise
X = vectorizer.fit_transform(permits["Cleaned_Description"])

# Perform clustering
n_clusters = 100  # Set the number of clusters based on expected categories
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
permits["Cluster"] = kmeans.fit_predict(X)
```

```python
# Assign a category to each cluster
permits["Category"] = permits["Cluster"].astype("category")
```

## Topic Modeling Using LDA

1. Transformed the text into a bag-of-words representation using CountVectorizer.

2. Applied LDA, which assigns probabilities to words for each topic and probabilities to topics for each document

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Vectorize text for LDA
vectorizer = CountVectorizer(max_features=5000, stop_words='english')
X_counts = vectorizer.fit_transform(permits['Cleaned_Description'])

# Apply LDA
lda = LatentDirichletAllocation(n_components=10, random_state=42)  # Adjust topics as needed
lda.fit(X_counts)

# Assign topic to each description
permits['Topic'] = lda.transform(X_counts).argmax(axis=1)
```

```python
# Get top keywords for each topic
for topic_idx, topic in enumerate(lda.components_):
    top_keywords = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[:-11:-1]]
    print(f"Topic {topic_idx}: {top_keywords}")
```

```python
# Assign meaningful labels to topics
topic_labels = {
    0: "Reroofing and Sprinkler Work",
    1: "Window Replacement and Repair",
    2: "Seismic Retrofit and Structural Work",
    3: "Permit Revisions and Plan Updates",
    4: "Roof and Stair Repairs",
    5: "Interior Remodels (Kitchen, Bathroom, Units)",
    6: "Maher Ordinance and Signage",
    7: "Interior Finishes and Partitions",
    8: "Street and Sidewalk Work",
    9: "General Work and Final Inspections",
}

# Map topic labels back to your dataset
permits['Topic_Label'] = permits['Topic'].map(topic_labels)

# Display the dataset with labels
print(permits[['Description', 'Topic', 'Topic_Label']].head())
```

```python
# Count the number of descriptions per topic
topic_counts = permits['Topic_Label'].value_counts()

# Plot topic distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
topic_counts.plot(kind='bar', color='skyblue')
plt.title("Topic Distribution")
plt.xlabel("Topics")
plt.ylabel("Number of Descriptions")
plt.xticks(rotation=45)
plt.show()
```
